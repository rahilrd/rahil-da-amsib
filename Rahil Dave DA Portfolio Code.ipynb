```python
#Task 7
#Creating Individual Scatter Plots

import pandas as pd
df = pd.read_csv("costofliving.csv")
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
col = df['Cost of Living Index'].to_numpy()
gro=df['Groceries Index'].to_numpy()
colr=df['Cost of Living Plus Rent Index'].to_numpy()
lpp=df['Local Purchasing Power Index'].to_numpy()
res=df['Restaurant Price Index'].to_numpy()

plt.scatter(col,gro)
z = np.polyfit(col, gro, 1)
p = np.poly1d(z)
plt.plot(col, p(col),'#09B016')
plt.xlabel("Cost of Living Index")
plt.ylabel("Groceries Index")
plt.title("Scatter Plot for Cost of Living vs Groceries")
plt.grid()
plt.show()

plt.scatter(col,res)
z = np.polyfit(col, res, 1)
p = np.poly1d(z)
plt.plot(col, p(col),'#09B016')
plt.xlabel("Cost of Living Index")
plt.ylabel("Restaurant Price Index")
plt.title("Scatter Plot for Cost of Living vs Restaurant Price")
plt.grid()
plt.show()

plt.scatter(col,res)
z = np.polyfit(col, res, 1)
p = np.poly1d(z)
plt.plot(col, p(col),'#09B016')
plt.xlabel("Cost of Living Index")
plt.ylabel("Restaurant Price Index")
plt.title("Scatter Plot for Cost of Living vs Restaurant Price")
plt.grid()
plt.show()
```

```python
#Task 7
#Creating a intial comparison for Cost of Living and Groceries with a trend line

import pandas as pd
df = pd.read_csv("costofliving.csv")
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import figure
from matplotlib.pyplot import axis
figure(figsize=(8,8))
col = df['Cost of Living Index'].to_numpy()
gro=df['Groceries Index'].to_numpy()
colr=df['Cost of Living Plus Rent Index'].to_numpy()
lpp=df['Local Purchasing Power Index'].to_numpy()
res=df['Restaurant Price Index'].to_numpy()

groceries=plt.scatter(col,gro,c='#4CEEF0')
z = np.polyfit(col, gro, 1)
p = np.poly1d(z)
grofit=plt.plot(col, p(col),'#3BB9BB')
plt.xlabel("Cost of Living Index",fontsize=15)
plt.ylabel("Groceries Index",fontsize=15)
plt.title("Scatter Plot for Cost of Living vs Groceries",fontsize=15)

plt.show()
```

```python
#Task 7
#Creating a multi index scatter plot to show multiple positive correaltions with trendlines and a legend

groceries=plt.scatter(col,gro,c='#4CEEF0')
restaurant=plt.scatter(col,res,c='#A04CF0')
purchasing=plt.scatter(col,lpp,c='#A2D940')
z = np.polyfit(col, gro, 1)
p = np.poly1d(z)
grofit=plt.plot(col, p(col),'#3BB9BB')
z = np.polyfit(col, res, 1)
p = np.poly1d(z)
resfit=plt.plot(col, p(col),'#9D5DB9')
z = np.polyfit(col, lpp, 1)
p = np.poly1d(z)
lppfit=plt.plot(col, p(col),'#5B7B22')
plt.xlabel("Cost of Living Index",fontsize=15)
plt.title("Scatter Plot for Cost of Living vs Other Indices",fontsize=15)
plt.legend((groceries,restaurant,purchasing),
           ("Groceries Index", "Restaurant Price Index","Local Purchasing Power Index"),
           scatterpoints=1,
           loc=(1.03,0.5),
           fontsize=15)

plt.show()
```
